```python
# CAPM CALCULATOR
# R = R_f + b(R_m - R_f)
R_f = float(input('Enter the risk free rate: '))
R_m = float(input('Enter the market rate: '))
b = float(input('Enter the beta: '))
R = R_f + b * (R_m - R_f)
print(format(R, ".3f"))
```
